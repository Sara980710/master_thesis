```python
import pandas as pd
```

```python
df = pd.read_csv("GlobalWatchData/trollers.csv")
```

```python
df.head()
```

```python
condition = (df["lat"] > 56.197894) & (df["lat"] < 57.541) & (df["lon"] > 9.931) & (df["lon"] < 12.720355)
kattegat_df = df[condition]
```

```python
kattegat_df.head(10)
```

```python
kattegat_df["timestamp"] = pd.to_datetime(kattegat_df["timestamp"], unit='s')
```

```python
kattegat_df.head()
```

```python
isFishing_df = kattegat_df[kattegat_df["is_fishing"] != -1]
isFishing_df.head()
```

```python
grouped_isFishing_df = isFishing_df.groupby("mmsi")
grouped_isFishing_df.first().head()
```

```python
noInfo_df = kattegat_df[kattegat_df["is_fishing"] == -1]
noInfo_df.head()
```

```python
grouped_noInfo_df = noInfo_df.groupby("mmsi")
grouped_noInfo_df.first().head()
```
