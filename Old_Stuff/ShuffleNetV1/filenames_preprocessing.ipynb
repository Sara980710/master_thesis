```python
import pandas as pd
```

```python
df = pd.read_csv("filenames.csv")
display(df.head())
```

```python
df_updated = df.replace(to_replace="[.jpg]", value="", regex=True)
df_updated.head()
```

```python
print(df_updated['folder_name'])
```

```python
for key, value in df_updated.iterrows():
    print(f'key: {key}, value: {value[0]}\n')
```

```python
with open('folderNames.txt', 'w') as f:
    for key, value in df_updated.iterrows():
        f.write(f'mkdir -p {value[0]}\n')
        f.write(f'mv {value[0]}.jpg {value[0]}\n')

f.close()
```
